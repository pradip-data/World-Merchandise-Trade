```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# 📂 Load dataset
file_path = r"C:\Users\chemi\Downloads\PROJECT -World Merchandise Trade  (Bigquery Project)\BigQuery Output Result\India's Global Trade Case Study (1948-2023).csv"
df = pd.read_csv(file_path)

# 🛠 Data Preprocessing
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df["Trade Balance"] = df["India_Exports"] - df["India_Imports"]
df["Total Trade"] = df["India_Exports"] + df["India_Imports"]

# 🔥 1. India's Exports & Imports Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="India_Exports", data=df, label="Exports", marker="o", color="blue")
sns.lineplot(x="Year", y="India_Imports", data=df, label="Imports", marker="s", color="red")
plt.title("India's Exports & Imports (1948-2023)")
plt.xlabel("Year")
plt.ylabel("USD Billion")
plt.legend()
plt.grid(True)
plt.show()

# 🔴 2. Trade Deficit Trend
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="Trade_Deficit", data=df, marker="o", color="darkred")
plt.axhline(0, color="black", linewidth=1)
plt.title("India's Trade Deficit Over Time")
plt.xlabel("Year")
plt.ylabel("Trade Deficit (USD Billion)")
plt.grid(True)
plt.show()

# 🌎 3. India's Share in Global Trade
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="India_Total_Trade_Percentage", data=df, marker="o", color="green")
plt.title("India's Share in Global Trade")
plt.xlabel("Year")
plt.ylabel("Percentage of Global Trade")
plt.grid(True)
plt.show()

# 🔥 4. Export & Import Growth Rates
plt.figure(figsize=(12, 6))
sns.lineplot(x="Year", y="Export_Growth_Percentage", data=df, label="Export Growth %", marker="o", color="blue")
sns.lineplot(x="Year", y="Import_Growth_Percentage", data=df, label="Import Growth %", marker="s", color="red")
plt.axhline(0, color="black", linewidth=0.5)
plt.title("Export & Import Growth Rates")
plt.xlabel("Year")
plt.ylabel("Growth (%)")
plt.legend()
plt.grid(True)
plt.show()




# 📄 Generate PDF Report
pdf_path = r"C:\Users\chemi\Downloads\India_Global_Trade_Report.pdf"
c = canvas.Canvas(pdf_path, pagesize=letter)
c.setFont("Helvetica-Bold", 14)
c.drawString(100, 750, "India's Global Trade - Detailed Analysis (1948-2023)")

c.setFont("Helvetica", 12)
c.drawString(100, 720, "1. Exports & Imports visualized over time.")
c.drawString(100, 700, "2. Trade Deficit trend analyzed.")
c.drawString(100, 680, "3. India's share in global trade displayed.")
c.drawString(100, 660, "4. Export & Import Growth Rates tracked.")

c.save()
print(f"✅ PDF Report Saved: {pdf_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# 📂 Load dataset
file_path = r"C:\Users\chemi\Downloads\PROJECT -World Merchandise Trade  (Bigquery Project)\BigQuery Output Result\India's Global Trade Case Study (1948-2023).csv"
df = pd.read_csv(file_path)

# 🛠 Data Preprocessing
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df["Trade Balance"] = df["India_Exports"] - df["India_Imports"]
df["Total Trade"] = df["India_Exports"] + df["India_Imports"]

# 🔍 Filter Data for 2023
df_2023 = df[df["Year"] == 2023]

# 🌟 5. India's Trade Breakdown (2023)
plt.figure(figsize=(8, 5))
labels = ["Exports", "Imports", "Trade Deficit"]
values = [df_2023["India_Exports"].values[0], df_2023["India_Imports"].values[0], abs(df_2023["Trade_Deficit"].values[0])]
colors = ["blue", "red", "darkred"]
plt.pie(values, labels=labels, autopct="%1.1f%%", colors=colors, startangle=140)
plt.title("India's Trade Breakdown - 2023")
plt.show()

# 📊 .6 India's Rank in Global Trade (2023)
plt.figure(figsize=(8, 5))
ranks = ["Export Rank", "Import Rank", "Trade Rank"]
rank_values = [df_2023["Export_Rank"].values[0], df_2023["Import_Rank"].values[0], df_2023["Trade_Rank"].values[0]]
sns.barplot(x=ranks, y=rank_values, palette="coolwarm")
plt.title("India's Rank in Global Trade - 2023")
plt.ylabel("Global Ranking")
plt.gca().invert_yaxis()  # Lower rank is better, so invert
plt.show()

# 📈 7. Export & Import Growth in 2023
plt.figure(figsize=(8, 5))
growth_labels = ["Export Growth %", "Import Growth %", "Total Trade Growth %"]
growth_values = [df_2023["Export_Growth_Percentage"].values[0], df_2023["Import_Growth_Percentage"].values[0], df_2023["Total_Trade_Growth_Percentage"].values[0]]
sns.barplot(x=growth_labels, y=growth_values, palette="viridis")
plt.title("India's Export & Import Growth (2023)")
plt.ylabel("Growth (%)")
plt.axhline(0, color="black", linewidth=1)
plt.show()

# 🔄 8. Trade Deficit: 2023 vs Previous Year
plt.figure(figsize=(8, 5))
years = ["2022", "2023"]
deficit_values = [df[df["Year"] == 2022]["Trade_Deficit"].values[0], df_2023["Trade_Deficit"].values[0]]
sns.barplot(x=years, y=deficit_values, palette="magma")
plt.title("India's Trade Deficit: 2023 vs 2022")
plt.ylabel("Trade Deficit (USD Billion)")
plt.show()

# 🌍 9. India's Share in Global Trade (2023 vs Previous Years)
plt.figure(figsize=(10, 6))
sns.lineplot(x="Year", y="India_Total_Trade_Percentage", data=df, marker="o", color="green")
plt.axvline(x=2023, color="red", linestyle="--", label="2023")
plt.title("India's Share in Global Trade Over Time")
plt.xlabel("Year")
plt.ylabel("India's Share in Global Trade (%)")
plt.legend()
plt.grid(True)
plt.show()

# 📄 Generate PDF Report with 2023 Insights
pdf_path = r"C:\Users\chemi\Downloads\India_Global_Trade_Report_2023.pdf"
c = canvas.Canvas(pdf_path, pagesize=letter)
c.setFont("Helvetica-Bold", 14)
c.drawString(100, 750, "India's Global Trade - 2023 Detailed Analysis")

c.setFont("Helvetica", 12)
c.drawString(100, 720, "1. India's Trade Breakdown (Exports, Imports, Trade Deficit).")
c.drawString(100, 700, "2. India's Global Rank in Exports, Imports, and Total Trade.")
c.drawString(100, 680, "3. India's Export & Import Growth in 2023.")
c.drawString(100, 660, "4. Trade Deficit Comparison: 2023 vs Previous Year.")
c.drawString(100, 640, "5. India's Share in Global Trade (Trend Analysis).")

c.save()
print(f"✅ PDF Report Saved: {pdf_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 📂 Load dataset
file_path = r"C:\Users\chemi\Downloads\PROJECT -World Merchandise Trade  (Bigquery Project)\BigQuery Output Result\India's Global Trade Case Study (1948-2023).csv"
df = pd.read_csv(file_path)

# 🛠 Data Preprocessing
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df["Trade Balance"] = df["India_Exports"] - df["India_Imports"]
df["Total Trade"] = df["India_Exports"] + df["India_Imports"]

# 🔍 Filter Data for 2023
df_2023 = df[df["Year"] == 2023]



# 🎯 10. Bubble Chart: Export vs Import Growth (Bubble = Trade Volume)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Export_Growth_Percentage"], y=df["Import_Growth_Percentage"], size=df["Total Trade"], sizes=(10, 500), hue=df["Year"], palette="coolwarm", edgecolor="black", alpha=0.7)
plt.xlabel("Export Growth (%)")
plt.ylabel("Import Growth (%)")
plt.title("Bubble Chart: Export vs Import Growth")
plt.axhline(0, color='black', linestyle="--")
plt.axvline(0, color='black', linestyle="--")
plt.grid(True)
plt.legend(title="Year", loc="best")
plt.show()

# 🎯 11. Stacked Area Chart: Trade Composition Over Time
plt.figure(figsize=(10, 6))
plt.stackplot(df["Year"], df["India_Exports"], df["India_Imports"], labels=["Exports", "Imports"], colors=["blue", "red"], alpha=0.6)
plt.title("Trade Composition Over Time")
plt.xlabel("Year")
plt.ylabel("USD Billion")
plt.legend()
plt.grid(True)
plt.show()

# 🎯 12. Histogram: Yearly Trade Deficit Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["Trade_Deficit"], bins=15, kde=True, color="darkred", alpha=0.7)
plt.axvline(df["Trade_Deficit"].mean(), color="black", linestyle="--", label="Avg Trade Deficit")
plt.title("Trade Deficit Distribution (All Years)")
plt.xlabel("Trade Deficit (USD Billion)")
plt.legend()
plt.grid(True)
plt.show()

# 🎯 13. India's Share in Global Trade (Donut Chart - 2023)
plt.figure(figsize=(6, 6))
sizes = [df_2023["India_Total_Trade_Percentage"].values[0], 100 - df_2023["India_Total_Trade_Percentage"].values[0]]
labels = ["India", "Rest of the World"]
colors = ["gold", "gray"]
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, startangle=140, wedgeprops=dict(width=0.3))
plt.title("India's Share in Global Trade (2023)")
plt.show()

# 🎯 14. Boxplot: Export & Import Growth Variations
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[["Export_Growth_Percentage", "Import_Growth_Percentage"]], palette="coolwarm")
plt.title("Export & Import Growth Variations Over Time")
plt.grid(True)
plt.show()

# 🎯 15. Bar Chart: Top 10 Years by Trade Growth %
top_years = df.nlargest(10, "Total_Trade_Growth_Percentage")
plt.figure(figsize=(10, 5))
sns.barplot(x=top_years["Year"], y=top_years["Total_Trade_Growth_Percentage"], palette="viridis")
plt.title("Top 10 Years by Trade Growth %")
plt.xlabel("Year")
plt.ylabel("Growth (%)")
plt.xticks(rotation=45)

# 📌 Add Data Labels
for index, value in enumerate(top_years["Total_Trade_Growth_Percentage"]):
    plt.text(index, value, f"{value:.2f}%", ha="center", va="bottom", fontsize=10, fontweight="bold", color="black")

plt.show()
```
